```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(2, 20, 10)
```

```python
m, c= 3, 10
```

```python
y= m*x + c
```

```python
plt.plot(x,y) #linear line
```

```python
plt.scatter(x,y)   #scatter points
```

```python
#adding 1000 points
```

```python
x = np.linspace(2, 20, 1000)
m, c= 3, 10
y= m*x + c
```

```python
plt.scatter(x,y)
```

```python
# 1000 points will make the dots a line (condensed points)
```

```python
#we using standard eviation because in randomly generated data, SD really realted to the real data quite a lot
```

```python
#noise
#we use noise because,
#if we do not have noice then we dont have real data, we've idealastic data
```

```python
noise= np.random.randn(1000)*5
```

```python
y_mod= y+ noise  #modified y
```

```python
plt.scatter(x,y_mod)
```

```python
#SEABORN
#seabord is a plotting library like matplotlib  but is used to plot in better scale
```

```python
import seaborn as sns
```

```python
x= np.random.randn(1000)
```

```python
sns.distplot(x, bins=30, kde=False, rug=True)
```

```python
data=sns.load_dataset("tips")   #loading dataset
```

```python
data.head()
```

```python
sns.scatterplot(data["total_bill"], data["tip"], hue=data["smoker"])
#hue is for color, here we represent how a person tip when he's smoker or non-smoker,
#hue takes column name for color encoding 
```

```python
type(data)
```

```python
sns.violinplot(y=data.total_bill)
```
